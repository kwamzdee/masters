```python
#Exercise 1 Stack
myStack=['a','b','c']

def stackAppend(option):
    myStack.append('d')
    print(myStack)

def stackPop(option):
    myStack.pop()
    print(myStack)

def stackView(option):
    myStack.copy()
    print(myStack)
    
def stackPeek(option):
    myStack.peek()
    print(myStack)

option = input ("Select an option")

while option !="exit":
    if option =="Push":
        stackAppend(option) 
    
    elif option =="Pop":
         stackPop(option)
    
    elif option == "View":
        stackView(option)
        
    else :
        print ("Invalid Option selected!Try Again")
    
    option = input("Select an Option")
```

```python
#Exercise 1 Queue

myQueue = ['a','b','e']

def queueEnqueue(option):
    myQueue.append('c')
    print(myQueue)

def queueDequeue(option):
    myQueue.pop(0)
    print(myQueue)

def queueView(option):
    myQueue.copy()
    print(myQueue)

option=input("Select an Option")

while option !="exit":
    
    if option =="Enqueue":
        queueEnqueue(option)

    elif option == "Dequeue":
        queueDequeue(option)
    
    elif option == "View":
         queueView(option)
    
    else :
        print("Invalid Option Entered!Try Again")
    
    option=input("Select an Option")
    

```
